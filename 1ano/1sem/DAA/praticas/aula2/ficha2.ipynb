```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import f1_score
from sklearn.metrics import fbeta_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#USING CROSS VALIDATION
from sklearn.model_selection import cross_val_score
```

```python
'''
Load dos datasets e merge de ambos por 'Area Code' e 'Phone'
'''
callsexcel = pd.read_excel('CallsData.xls')
contractdatacsv = pd.read_csv('ContractData.csv')

allData = callsexcel.merge(contractdatacsv,how="inner",on=["Area Code","Phone"])
```

```python

#allData_wbinary = pd.get_dummies(allData, columns=['Churn']) converter categorico em binario
allData['Churn'] = allData['Churn'].map({0: 'Permaneceu', 1: 'Abandonou'})
allData
```

```python
X = allData.drop(['Churn'],axis=1)
Y = allData['Churn'].to_frame()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=12345) #random_state->seed

X_train = X_train.drop(['State','Phone'], axis=1)  
X_test = X_test.drop(['State','Phone'], axis=1)    

clf = DecisionTreeClassifier(random_state=12345)

clf.fit(X_train, Y_train)
predictions = clf.predict(X_test)
```

```python
accuracy_score(Y_test, predictions)
```

```python
f1_score(Y_test, predictions, average='macro')
```

```python
from sklearn.model_selection import KFold
scores = []
i=0
kf = KFold(n_splits=10)
for train, test in kf.split(X): 
    clf.fit(X.loc[train,:],Y.loc[train,:])
    score = clf.score(X.loc[test,:],Y.loc[test,:])
    scores.append(score)
    y_predicted = clf.predict(X.loc[test,:])
    print("Confusion Matrix:")
    print(confusion_matrix(Y.loc[test,:], y_predicted))
    print(score)
print("RESULT: %0.2f accuracy with a standard deviation of %0.2f" % (np.mean(scores),np.std(scores)))
```
